```python
# Uninstall conflicting versions first
!pip uninstall -y transformers
!pip install transformers==4.36.2 rouge-score==0.1.2 gdown
from transformers import T5Tokenizer, T5ForConditionalGeneration
from rouge_score import rouge_scorer
import torch
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
def t5_summarize(text):
    inputs = tokenizer("summarize: " + text, return_tensors="pt", max_length=512, truncation=True).to(device)
    summary_ids = model.generate(
        inputs['input_ids'],
        max_length=150, min_length=40,
        length_penalty=2.0, num_beams=4,
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
def compute_rouge(summary, reference):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, summary)
    return scores
text = """
The Document Academy is a platform for researchers to explore documents as communicative artifacts in society.
In recent years, interest has increased in understanding the meaning and function of documents in both historical and contemporary settings.
"""

# Generate summary
summary = t5_summarize(text)
print("Generated Summary:\n", summary)

# Compute ROUGE scores against a reference summary
reference_summary = "The Document Academy explores the role and meaning of documents in society, historically and today."
scores = compute_rouge(summary, reference_summary)
print("ROUGE Scores:\n", scores)

```
